## Task 1: A List of Dictionaries

```python
transactions = [
    {"id": 1, "product": "Laptop",   "price": 1200, "quantity":  2, "region": "EU"},
    {"id": 2, "product": "Phone",    "price":  800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet",   "price":    0, "quantity":  5, "region": "EU"},
    {"id": 4, "product": "Monitor",  "price":  350, "quantity":  3, "region": "US"},
    {"id": 5, "product": "Laptop",   "price": 1200, "quantity":  1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price":   75, "quantity": 10, "region": "APAC"},
]

# a) Filter valid transactions
valid_transactions = [
    t for t in transactions
    if t["price"] > 0 and t["quantity"] >= 0
]

# b) Add total field
for t in valid_transactions:
    t["total"] = t["price"] * t["quantity"]

# c) Group by region
revenue_by_region = {}
for t in valid_transactions:
    region = t["region"]
    revenue_by_region[region] = revenue_by_region.get(region, 0) + t["total"]

print("Revenue by region:", revenue_by_region)

# d) Find top region
top_region = max(revenue_by_region, key=revenue_by_region.get)
print(f"Top region: {top_region} with revenue {revenue_by_region[top_region]}")
```

## Task 2: Creating a Function

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

def summarise_sales(daily_sales):
    """Return a summary dictionary for a list of daily sales values."""
    total = sum(daily_sales)
    average = round(total / len(daily_sales), 2)
    best_day = daily_sales.index(max(daily_sales)) + 1   # 1-based index
    worst_day = daily_sales.index(min(daily_sales)) + 1  # 1-based index
    above_average_count = sum(1 for s in daily_sales if s > average)

    return {
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average_count,
    }

summary = summarise_sales(sales)
print(summary)
```

## Task 3: Loops

```python
students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia",  "grade": 91, "attendance": 0.98},
    {"name": "Leo",  "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom",  "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]

# a) Students with grade >= 50
passing = [s["name"] for s in students if s["grade"] >= 50]
print("Passing students:", passing)

# b) Status message for each student
print()
for s in students:
    if s["grade"] >= 80:
        status = "Excellent"
    elif s["grade"] >= 50:
        status = "Pass"
    else:
        status = "Fail"
    print(f"{s['name']}: {status}")

# c) Count students with attendance below 0.80
low_attendance_count = sum(1 for s in students if s["attendance"] < 0.80)
print(f"\nStudents with attendance below 0.80: {low_attendance_count}")

# d) Find highest grade using a loop
top_student = students[0]
for s in students[1:]:
    if s["grade"] > top_student["grade"]:
        top_student = s
print(f"\nTop student (loop): {top_student['name']} with grade {top_student['grade']}")

# e) Find highest grade using max()
best = max(students, key=lambda s: s["grade"])
print(f"Top student (max): {best['name']} with grade {best['grade']}")
```

## Task 4: Class and Methods

```python
class Portfolio:
    """Represents an investment portfolio with multiple asset holdings."""

    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}

    def add_asset(self, ticker, shares, price_per_share):
        """Add a new asset or update shares and price for an existing ticker."""
        if ticker in self.holdings:
            self.holdings[ticker]["shares"] += shares
        else:
            self.holdings[ticker] = {"shares": shares}
        # Always update the price to the latest value
        self.holdings[ticker]["price_per_share"] = price_per_share

    def total_value(self):
        """Return the total market value of the portfolio."""
        return sum(
            data["shares"] * data["price_per_share"]
            for data in self.holdings.values()
        )

    def most_valuable(self):
        """Return the ticker with the highest individual asset value."""
        return max(
            self.holdings,
            key=lambda t: self.holdings[t]["shares"] * self.holdings[t]["price_per_share"]
        )

    def __str__(self):
        return (
            f"Portfolio of {self.owner} | "
            f"Assets: {len(self.holdings)} | "
            f"Total value: €{self.total_value():,.2f}"
        )


# Demonstration
p = Portfolio("Alice")
p.add_asset("AAPL", 10, 185.50)
p.add_asset("MSFT", 5,  420.00)
p.add_asset("NVDA", 8,  870.00)

# Add more shares to an existing ticker to show update logic
p.add_asset("AAPL", 5, 190.00)

print(p)
print(f"Total portfolio value: €{p.total_value():,.2f}")
print(f"Most valuable asset: {p.most_valuable()}")
```

## Task 5: Importing a Module

```python
from datetime import date

orders = [
    ("2025-01-15", "Laptop"),  ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"),  ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"),   ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"),  ("2025-04-05", "Monitor"),
]

# Parse each date string into a date object
order_dates = [date.fromisoformat(order_date) for order_date, _ in orders]

earliest = min(order_dates)
latest = max(order_dates)
days_between = (latest - earliest).days

print(f"Earliest order: {earliest}")
print(f"Latest order:   {latest}")
print(f"Days between earliest and latest order: {days_between}")
```

## Task 6: Array Operations and Broadcasting

```python
import numpy as np

# Rows: years (2022, 2023, 2024), Columns: quarters (Q1, Q2, Q3, Q4)
revenue = np.array([
    [210, 230, 190, 250],
    [220, 260, 200, 270],
    [250, 290, 230, 310]
])

# a) Total annual revenue per year (row sums)
annual_totals = revenue.sum(axis=1)
print("Annual totals (2022–2024):", annual_totals)

# b) Mean quarterly revenue across all years (column means)
quarterly_means = revenue.mean(axis=0)
print("Mean quarterly revenue (Q1–Q4):", quarterly_means)

# c) Apply growth adjustment to the 2024 row via broadcasting
growth_factors = np.array([1.05, 1.03, 1.08, 1.02])
adjusted_2024 = revenue[2] * growth_factors
print("Adjusted 2024 revenue:", adjusted_2024)

# d) All quarterly values exceeding 250
above_250 = revenue[revenue > 250]
print("Quarterly revenues above 250:", above_250)
```

## Task 7: Pandas

```python
import pandas as pd

data = {
    "employee":          ["Anna", "Erik", "Mia", "Leo", "Sara", "Tom"],
    "department":        ["Finance", "Marketing", "Finance", "IT", "Marketing", "IT"],
    "salary":            [4500, 4000, 4800, 5200, 3900, 5100],
    "years_exp":         [5, 3, 7, 10, 2, 8],
    "performance_score": [4.2, 3.8, 4.5, 4.0, 3.5, 4.9],
}
df = pd.DataFrame(data)

# a) Add salary_per_year_exp column
df["salary_per_year_exp"] = (df["salary"] / df["years_exp"]).round(0).astype(int)

# b) Top 3 employees by performance score
top3 = df.sort_values("performance_score", ascending=False).head(3)
print("Top 3 by performance score:")
print(top3[["employee", "performance_score"]], "\n")

# c) Finance employees with performance score > 4.0
finance_high_performers = df.loc[
    (df["department"] == "Finance") & (df["performance_score"] > 4.0)
]
print("Finance employees with score > 4.0:")
print(finance_high_performers, "\n")

# d) Rename years_exp to experience_years
df = df.rename(columns={"years_exp": "experience_years"})

# e) Salary and performance_score for first four employees (via iloc)
print("Salary & performance for first 4 employees:")
print(df.iloc[:4][["salary", "performance_score"]])
```

## Task 8: More Pandas

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors         = rng.integers(200, 600, size=28)
conversion_rate  = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value  = rng.normal(50, 10, size=28).round(2)

# a) Create DataFrame with date as index
traffic = pd.DataFrame(
    {
        "visitors":        visitors,
        "conversion_rate": conversion_rate,
        "avg_order_value": avg_order_value,
    },
    index=days
)
traffic.index.name = "date"

# b) Compute conversions and daily_revenue
traffic["conversions"]    = (traffic["visitors"] * traffic["conversion_rate"]).round(0).astype(int)
traffic["daily_revenue"]  = traffic["conversions"] * traffic["avg_order_value"]

print(traffic.head(10).to_string())
```
